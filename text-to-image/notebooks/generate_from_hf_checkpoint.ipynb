```python
from micro_diffusion.models.model import create_latent_diffusion
from huggingface_hub import hf_hub_download
import torch

# Init model
params = {
    'latent_res': 64,
    'in_channels': 4,
    'pos_interp_scale': 2.0,
    'dtype': 'float32',
}
model = create_latent_diffusion(**params).to('cuda:1')
```
